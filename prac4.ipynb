```python
#grading system
def grading_system(mark):

    if mark > 75:
       return 'Distinction'
    elif mark >= 50:
        return 'Pass'
    else:
        return 'Sorry but it is a fail'

mark = float(input('Enter Mark:'))
result = grading_system(mark)

print(f'Grade is {mark}% , {result}')

 
```
